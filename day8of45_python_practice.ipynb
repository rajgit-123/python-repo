```python
#working with groupby and aggregations
```

```python
import pandas as pd
sales=pd.DataFrame({
    'weekday' :['Sun','Sun','Mon','Mon'],
    'city':['Austin','Dallas','Austin','Dallas'],
    'bread':[139,237,326,456],
    'butter':[20,45,70,98]

})
sales
```

```python
#boolean filter count
sales.loc[sales['weekday']=='Sun'].count()

#groupby Count
sales.groupby('weekday').count()

#groupby and sum
sales.groupby('weekday')['bread'].sum()

sales.groupby('weekday')['butter'].sum()
#group by and sum on multiple columns
sales.groupby('weekday')[['bread','butter']].sum()

#groupny and mean :multilevel index
sales.groupby(['city','weekday']).mean()
#sales.groupby(['bread','weekday']).mean()

#groupby and max on multiple coulmns
sales.groupby('weekday')[['bread','butter']].max()
#sales.groupby(['city','weekday']).max()

##multiple aggregations
sales.groupby('city')[['bread','butter']].agg(['max','min'])

#custom aggregation
def data_range(series):
  return series.max()-series.min()
sales.groupby('weekday')[['bread','butter']].agg(data_range)

```
