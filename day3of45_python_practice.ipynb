```python
#create list of strings
names=['ram','rabson','robert','nancy']
#print each list itesm in names using for loop
for name in names:
  print(name)
#2, replicate for loop output with list_iterator output
names_iter=names.__iter__()
print(type(names_iter))
print(next(names_iter))
print(next(names_iter))
print(next(names_iter))
print(next(names_iter))
```

```python
#range iterator
```

```python
#1, loop over range(5) and print output
for num in range(5):
  print(num)

#2,create iterator
range_iter=iter(range(5))
print(type(range_iter))
print(next(range_iter))
print(next(range_iter))
print(next(range_iter))
print(next(range_iter))
print(next(range_iter))
```
