```python
#SCATTERPLOT CHART
import matplotlib.pyplot as plt
import numpy as np

year = [1950,1970,1990,2010]
pop = [2.519,3.692,5.263,6.972]

np_pop=np.array(pop)
np_pop=np_pop*20
plt.scatter(year,pop,s=np_pop)
plt.xlabel("Year")
plt.ylabel("Population")
plt.title("Year wise Population")
plt.text(1950,2.519,'India')
plt.text(1970,3.692,'China')
plt.show()
```

```python
#MULTPLE PLOTS ON SIGLE AXIS

import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "/percent-bachelors-degrees-women-usa.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "yogeshkumarsingla/percentage-of-bachelors-degrees-done-by-women-usa",
  file_path
)

print("First 5 records:", df.head())
```

```python
year=df['Year']
physical_sciences=df['Physical Sciences']
computer_science=df['Computer Science']
health=df['Health Professions']
education=df['Education']

import matplotlib.pyplot as plt
plt.plot(year,physical_sciences,color='blue',label='Physical Sciences')
plt.plot(year,computer_science,color='red',label='Computer Science')
plt.plot(year,health,color='green',label='Health Professions')
plt.plot(year,education,color='yellow',label='Education')
plt.plot(year,df['Agriculture'],color='black',label='Agriculture')
plt.plot(year,df['Engineering'],color='orange',label='Engineering')
plt.plot(year,df['Psychology'],color='purple',label='Psychology')
plt.plot(year,df['Public Administration'],color='pink',label='Public Administration')
plt.plot(year,df['Social Sciences and History'],color='brown',label='Social Sciences and History')



#plot all coumns with different colors
plt.legend(loc='best')
plt.show()


plt.subplot(1,2,1)
plt.plot(year,physical_sciences,color='blue',label='Physical Sciences')
plt.plot(year,computer_science,color='red',label='Computer Science')
plt.plot(year,health,color='green',label='Health Professions')
plt.tight_layout()
plt.legend(loc='best')
plt.show()

plt.subplot(2,2,4)
plt.plot(year,physical_sciences,color='blue',label='Physical Sciences')

plt.xlim(1990,2010)
plt.ylim(0,50)

plt.axis([1990,2010,0,50])
plt.tight_layout()
plt.legend(loc='lower right')
plt.show()

cs_max=df['Computer Science'].max()
#print(cs_max)
yr_max=year[computer_science.argmax()]
#print(cs_min)
plt.annotate('Maximum',xy=(yr_max,cs_max),xytext=(yr_max+5,cs_max+5),arrowprops=
             dict(facecolor='black'))
plt.show()

plt.style.use('ggplot')
plt.subplot(1,2,1)
plt.plot(year,physical_sciences,color='blue',label='Physical Sciences')
plt.plot(year,computer_science,color='red',label='Computer Science')
plt.plot(year,health,color='green',label='Health Professions')
plt.tight_layout()


```
