```python
#Calculate BMI using list
height=[1.70,1.59,1.51,1.73,1.92]
weight=[65.9,58.8,66.8,83.4,68.7]
#BMI=weight/height **2   #TypeError: unsupported operand type(s) for ** or pow(): 'list' and 'int'

#print(BMI)

#BMI Using numpy
import numpy as np
np_height=np.array(height)
np_weight=np.array(weight)
print(np_height)
print(np_weight)

BMI=np_weight/np_height ** 2
print(BMI)
print(BMI[1])
print(BMI>23)
print(BMI[BMI>23])
```
