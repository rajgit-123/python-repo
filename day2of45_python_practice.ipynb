```python
#nested function
def outer():
    n=1
    def inner():
        nonlocal n
        n=2
        print(n)
    inner()
    print(n)
outer()
```

```python
#Default arguments
def power(number, pow=2):
  new_val=number **pow
  return new_val
print(power(3))
print(power(3,3))
```

```python
#flexible arguments
def add_all(*args):
  sum_all=0
  for num in args:
    sum_all =sum_all+num
  return sum_all
print(add_all(2,3))
print(add_all(10,20,30,40,50))
```

```python
#flexible arguments kwargs
def print_all(**kwargs):
  for key, value in kwargs.items():
    print(key + ": " + value)
print_all(student1="ram",student2="sam")
```

```python
#LAMDA FUNCTIONS
```

```python
#simple defination for power function
def raise_to_poer(x,y):
  return x** y
raise_to_poer(2,3)

#same defination with lamda function
raise_to_power=lambda x,y:x**y
raise_to_power(2,3)

#regular defination for filter function
nums=[3,6,4,10,2,7,1]
def more_than_5(x):
  y=[]
  for i in x:
    if i>5:
      y.append(i)
  return y
more_than_5(nums)

#same defination with lamda function
more_than_5=filter(lambda x:x>5, nums)
list(more_than_5)
```
