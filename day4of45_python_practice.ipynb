Use iterator as function

```python
range_object=range(10,21)
print(type(range_object))
print(range_object)

#use this iterator object as a paramneter of list, max functions
#create a list of integers:range_object_list
range_object_list=list(range_object)
print(range_object_list)

#get max of values:range_object_list_max
range_object_list_max=max(range_object_list)
print(range_object_list_max)

```

```python
#ENUMARATION FUNCTION
DataScience=['PYTHON','R','MATHS','STATS','ML','DL','NLU','NLP','AI']
print(DataScience)
#1,enumarate() function creates enumate object
DataScience_enum=enumerate(DataScience)
print(type(DataScience_enum))
#2,create list of tuples
DataScience_enum_list=list(DataScience_enum)
print(DataScience_enum_list)
#3,unpack and print tuples
for index1,values1 in enumerate(DataScience):
    print(index1,values1)
#4,unpack and print with given start index
for index4,values4 in enumerate(DataScience,start=4):
  print(index4,values4)
```

```python
#ZIP FUnction
#create a list of tuples
std_no=(10,20,30,40)
std_name=('ram','bhim','nancy','rabinson')
std_marks=(100,200,300,400)

#Create a ZIP object
std_data=zip(std_no,std_name,std_marks)
print(type(std_data))
print(list(std_data))

#unpack zip and print tuple values
std_data1=zip(std_no,std_name,std_marks)
for no,name,marks in std_data1:
  print(no,name,marks)
#create zip with names and marks
z1=zip(std_name,std_marks)
print(list(z1))
print(*z1)  # this is to unpack to tuples from zip
```

```python
#LIST COMPREHENSIONS to avoid loop or LAMDA functions
nums=[10,20,30,40]
new_nums=[]
for num in nums:
  new_nums.append(num**2)
new_nums
#Using lambda functions
a=map(lambda x:x**2,nums)
list(a)
#Using list comprehensions
new_nums1=[num **2 for num in nums]
new_nums1

#List comprehensions to AVOID nested loops
pairs_1=[]
for num1 in range(0,2):
  for num2 in range(6,8):
    pairs_1.append((num1,num2))
print(pairs_1)
#nested loop using list comprehension
pairs2=[(num1,num2) for num1 in range(0,2) for num2 in range(6,8)]
print(pairs2)

#list comprehension with if condition
print([num**2 for num in range(10) if num%2==0])

#positive and negetive numbers
pos_neg=[{num:-num for num in range(10)}]
print(pos_neg)
print(type(pos_neg))
```

```python
# GENERATORS--Generator function that allows to you to declared a function that
# behaves like an interator, i.e it can be used in a for loop
l=[n*2 for n in range(1000)]  #list comprehension
g=(n*2 for n in range(1000))  #Generator Expression
print(type(l))
print(type(g))
import sys
print(sys.getsizeof(l))
print(sys.getsizeof(g))

print(l[4])
#print(g[4])  #TypeError: 'generator' object is not subscriptable
for v in l:
  print(v)
for v in g:
  print(v)
```
